```python
catalog = 'operations'
schema = 'finance_staging'
table_name = 'raw_tag_tbl'
```

```python
with open("sql/raw_tag_tbl.sql", "r") as file:
    qry = file.read()
(
    spark.sql(qry)
    .write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable(f"{catalog}.{schema}.{table_name}")
)
```
